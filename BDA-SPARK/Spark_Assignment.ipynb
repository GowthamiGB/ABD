```python
# 10. Map the  Registration  numbers to corresponding  branch. 58000 series  BDA,57000
# series AIML, 38000 series VLSI, 39000 series ES, and47000 series CDC.Given registration number,
# generate a key-value pair of Registration Number and Corresponding Branch.
def branch(d):
  if '580' in d:
    return 'BDA'
  elif '570' in d:
    return 'AIML'
  elif '380' in d:
    return 'VLSI'
  elif '390' in d:
    return 'ES'
  elif '470' in d:
    return 'CDC'
```

```python
students.flatMap(lambda d:d.split(' ')).map(lambda d:(d,branch(d))).collect()
```

```python
# 11.Text file contain numbers.
#  Numbers are separated by one white space. There is no order to store the numbers. One line may contain one or more numbers.
#  Find the maximum, minimum, sum and mean of numbers
num=spark.sparkContext.textFile('drive/My Drive/BDA-SPARK/number.txt')
num.collect()
```

```python
max=num.flatMap(lambda d:d.split(' ')).filter(lambda d:len(d)>0).map(lambda d:int(d)).max()
min=num.flatMap(lambda d:d.split(' ')).filter(lambda d:len(d)>0).map(lambda d:int(d)).min()
sum=num.flatMap(lambda d:d.split(' ')).filter(lambda d:len(d)>0).map(lambda d:int(d)).sum()
mean=num.flatMap(lambda d:d.split(' ')).filter(lambda d:len(d)>0).map(lambda d:int(d)).mean()
print(max,min,sum,mean)
```

```python

# 12. A text file (citizen.txt) contains data about citizens of country. Fields(information in file) are Name,
# dob, Phone, email and state name. Another file contains mapping of state names to state code
# like Karnataka is codes as KA, TamilNadu as TN, Kerala KL etc. Compress the citizen.txt file by
# changing full state name to state code.


stateRDD=spark.sparkContext.textFile('drive/My Drive/BDA-SPARK/codes.txt')
stateKey=stateRDD.map(lambda d:(d.split(' ')[0],d.split(' ')[1]))
stateKey.collect()
codes_dict={}
for val in stateKey.collect():
  codes_dict[val[0]]=val[1]
codes_dict



def changeToCode(d,codes_dict):
  new_data=' '
  split_data=d.split(' ')
  split_data[4]=codes_dict.get(split_data[4])
  new_data=new_data.join(split_data)
  return new_data


citizens=spark.sparkContext.textFile('drive/My Drive/BDA-SPARK/citizens.txt')
mapCitizens=citizens.map(lambda d:changeToCode(d,codes_dict))
mapCitizens.collect()
#mapCitizens.repartition(1).saveAsTextFile('/modified_citizens.txt') #Saving the updated file in RDD
```
